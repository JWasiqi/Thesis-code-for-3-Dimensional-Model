```python
p = var('p') #P
i = var('i') #I
a = var('a') #T
kP = var('kP') #kp
kIP = var('kIP') #kip
kPI = var('kPI') #kpi
kI = var('kI') #ki
kAI = var('kAI') #kai
kIA = var('kIA') #kia
kA = var('kA') #ka
```

```python
eq1 = p*(kP*(1-p) - kIP*i);
eq2 = i*(kPI*p- kI - kAI*a);
eq3 = a*(kIA*i - kA);
solutions = solve([eq1==0,eq2==0,eq3==0],p,i,a);
```

```python
J = jacobian([eq1,eq2,eq3],[p,i,a]);J
```

```python
A1 = kA - kIA*kP/kIP + kI*kP/kPI + kI*kIA*kP/(kIP*kPI);

A1 = A1.factor().substitute(kA == 1, kIA == 1); A1
```

```python
A2 = (kI*kP + kI^2*kIA*kP^2/(kIP*kPI^2) + kA*kI*kP/kPI - kI^2*kP/kPI - kI*kIA*kP^2/(kIP*kPI))
A2 = A2.factor().substitute(kA == 1, kIA == 1); A2
```

```python
A2n = (kI*kIA*kP + kA*kIP*kPI - kI*kIP*kPI - kIA*kP*kPI + kIP*kPI^2)*kI*kP;
A2n = A2n.full_simplify(); A2n
```

```python
A3 = kA*kI*kP - kI*kIA*kP^2/kIP - kI^3*kIA*kP^2/(kIP*kPI^2) - kA*kI^2*kP/kPI + 2*kI^2*kIA*kP^2/(kIP*kPI);
A3 = A3.factor().substitute(kA == 1, kIA == 1); A3
```
